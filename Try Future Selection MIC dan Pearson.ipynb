```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from minepy import MINE
```

```python
df = pd.read_csv("datakipasnew.csv") 
df
```

```python
df_zone1 = df.loc[(df["year"] == 2021 )]
df_zone1.info()
df_zone1
```

```python
df_factors1 = df_zone1.loc[:,["year","month","day","voltage","current","power","frequency","powerfactor","Energy"]].iloc[1:,:]
df_factors1.index=range(0,len(df_factors1))
df_factors1
```

```python
df_factors2 = df_zone1.loc[:,["Energy"]].iloc[:-1,:]
df_factors2.columns=["energy all"]
df_factors2.index=range(0,len(df_factors2))
df_factors2
```

```python
df_factors3 = df_factors1.join(df_factors2).loc[:,["year","month","day","voltage","current","power","frequency","powerfactor","Energy"]].iloc[1:,:]
df_factors3.info()
df_factors3
```

```python
df_factors4 = df_factors3.loc[:,["voltage"]].iloc[1:,:]
df_factors4.index=range(0,len(df_factors4))
df_factors4.columns=["voltageall"]
df_factors4.info()
df_factors4
```

```python
df_factors5 = df_factors3.loc[:,["current"]].iloc[1:,:]
df_factors5.index=range(0,len(df_factors4))
df_factors5.columns=["currentall"]
df_factors5.info()
df_factors5
```

```python
df_factors6 = df_factors3.loc[:,["power"]].iloc[1:,:]
df_factors6.index=range(0,len(df_factors4))
df_factors6.columns=["powerall"]
df_factors6.info()
df_factors6
```

```python
df_factors7 = df_factors3.loc[:,["frequency"]].iloc[1:,:]
df_factors7.index=range(0,len(df_factors4))
df_factors7.columns=["frequencyall"]
df_factors7.info()
df_factors7
```

```python
df_factors8 = df_factors3.loc[:,["powerfactor"]].iloc[1:,:]
df_factors8.index=range(0,len(df_factors4))
df_factors8.columns=["powerfactorall"]
df_factors8.info()
df_factors8
```

```python
df_factors9 = df_factors3.loc[:,["Energy"]].iloc[1:,:]
df_factors9.index=range(0,len(df_factors4))
df_factors9.columns=["Energyall"]
df_factors9.info()
df_factors9
```

```python
x1 = df_factors9["Energyall"]
y1 = df_factors4["voltageall"]
mine = MINE(alpha = 0.6, c = 15) 
mine.compute_score(x1, y1)
mic1 = mine.mic() #0.41602
pearson1 = np.corrcoef(x1, y1)[0, 1]
fig1 = plt.figure(figsize=(8,6))
ax1 = fig1.add_subplot(1,1,1)
ax1.scatter(x1, y1)
ax1.set_title("Pengaruh Energy thd Voltage")
ax1.set_ylabel('Voltage')
ax1.set_xlabel('Energy (KHW)')
print("MIC:", mic1)
print("Pearson:",pearson1)
```

```python
x2 = df_factors9["Energyall"]
y2 = df_factors5["currentall"]
mine = MINE(alpha = 0.6, c = 15) 
mine.compute_score(x2, y2)
mic2 = mine.mic() #0.41602
pearson2 = np.corrcoef(x2, y2)[0, 1]
fig2 = plt.figure(figsize=(8,6))
ax2 = fig2.add_subplot(1,1,1)
ax2.scatter(x2, y2)
ax2.set_title("Pengaruh Energy thd Current")
ax2.set_ylabel('Current')
ax2.set_xlabel('Energy (KHW)')
print("MIC:", mic2)
print("Pearson:",pearson2)
```
